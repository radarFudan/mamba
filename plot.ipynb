```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_ref = pd.read_csv('mamba_ssm_ref.csv')
df_ref_speed = pd.read_csv('mamba_ssm_ref_speedup.csv')
df_cuda = pd.read_csv('mamba_ssm_cuda.csv')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6), dpi=300)
ax.plot(df_ref['Length'], df_ref['Time'], label='Ref (in PyTorch)')
ax.plot(df_ref_speed['Length'], df_ref_speed['Time'], label='Ref Speedup (in PyTorch)')
ax.plot(df_cuda['Length'], df_cuda['Time'], label='Mamba (in CUDA)')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Forward Length')
plt.ylabel('Time (s)')
plt.title('Repeated 3 times')
plt.savefig('speedup_ref_with_associative_scan.pdf')
plt.legend()
```
